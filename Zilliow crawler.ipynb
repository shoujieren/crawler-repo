```python
import numpy as np 
import pandas as pd 
import re
import requests
from bs4 import BeautifulSoup
```

```python
headers={"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/108.0.0.0 Safari/537.36","Accept-Language":"en-US,en;q=0.9,zh-CN;q=0.8,zh;q=0.7","Accept":"*/*","Accept-Encoding":"gzip, deflate, br"}
```

```python
price_list=[]
size_list=[]
address_list=[]
```

```python
for page in range (1,10):
    
    resp= requests.get('https://www.zillow.com/homes/Charlotte,-NC_rb/'.format(page), headers = headers).text

    soup = BeautifulSoup(resp, 'html.parser')

    properties = soup.find_all("div",{"class":"StyledPropertyCardDataWrapper-c11n-8-73-8__sc-1omp4c3-0 gXNuqr property-card-data"})
    for x in range(0,len(properties)):
        try:
            price= properties[x].find("div",{"class":"StyledPropertyCardDataArea-c11n-8-73-8__sc-yipmu-0 hRqIYX"}).text
    
            size = properties[x].find("div",{"class":"StyledPropertyCardDataArea-c11n-8-73-8__sc-yipmu-0 ghGYOB"}).text

            address= properties[x].find("a",{"class":"StyledPropertyCardDataArea-c11n-8-73-8__sc-yipmu-0 lhIXlm property-card-link"}).text
        except:
            pass
price_list.append(price)
size_list.append(size)
address_list.append(address)
        
```

```python
df=pd.DataFrame()
```

```python
df['price']=price_list
df['size']=size_list
df['address']=address_list
df
```
